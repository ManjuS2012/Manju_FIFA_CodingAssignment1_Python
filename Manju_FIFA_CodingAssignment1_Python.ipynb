# Data cleaning and preprocessing

```python
# Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix.
# - Note: When the record/entry has "M"(indicates millions) as suffix need to multiply that value with 1000000
# - When the record/entry has "K"(indicates thousands) as suffix need to multiply that value with 1000

def fun_convert(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1]
        
        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value
fifa_df4['Value']=fifa_df4['Value'].apply(fun_convert)
fifa_df4['Wage'] = fifa_df4['Wage'].apply(fun_convert)
fifa_df4
```

```python
# Convert the column "Release Clause" to float datatype after getting rid of currency symbol and suffix.
# - Note: When the record/entry has "M"(indicates millions) as suffix need to multiply that value with 1000000
# - When the record/entry has "K"(indicates thousands) as suffix need to multiply that value with 1000

def fun_convert(df_value):
    
    if(pd.isna(df_value))!= True:
        try:
            value = float(df_value[1:-1])
            suffix = df_value[-1]

            if suffix == 'M':
                value = value * 1000000
            elif suffix == 'K':
                value = value * 1000
        except ValueError:
            value = 0
        return value
    else:
        return df_value

fifa_df4['Release Clause']=fifa_df4['Release Clause'].apply(fun_convert)
fifa_df4
```

```python
# Convert the column "Joined" into integer data type with keeping only the year.
fifa_df4['Joined'].isna().sum()
```

```python
# fifa_df4['Joined']=fifa_df4['Joined'].fillna('0')
```

```python
def str4(x):
    if isinstance(x,str):
        return int(x[-4:])
```

```python
fifa_df4['Joined'] = fifa_df4['Joined'].apply(str4)
fifa_df4['Joined'].head(5)
```

```python
# Convert the column "Contract Valid Until" to pandas datetime type.
fifa_df4['Contract Valid Until']=pd.to_datetime(fifa_df4['Contract Valid Until'])
fifa_df4['Contract Valid Until']
```

```python
# The column 'Height' is in inches with a quotation mark, Convert to float with decimal points. 
def feet(x):
    l=[]
    if isinstance(x,str):
        l = x.split("'")
        i = int(l[0])
        j = int(l[1])
        f = ((i*12)+j)/12
        return f
fifa_df4['Height'] = fifa_df4['Height'].apply(feet) 
```

```python
fifa_df4['Height'].head()
```

```python
# The column "Weight" has the suffix as lbs, remove the suffix and convert to float. 
def llbs(x):
    l=[]
    if isinstance(x,str):
        l = x.split("lbs")
        i = l[0]
        return float(i)
fifa_df4['Weight'] = fifa_df4['Weight'].apply(llbs)
```
